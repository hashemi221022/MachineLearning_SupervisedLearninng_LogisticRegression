# Normalize the data frame

```python
# normalized
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
normal = scaler.fit_transform(df)
df_normal = pd.DataFrame(normal, columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'])
```

# Separate independent variables from dependent variables

```python
# create X anh Y
X = pd.DataFrame(df_normal, columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
y = df_normal.Outcome.values.reshape(-1,1)
```

 # Create test and train data

```python
# create train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# create the model

```python
# create model
logreg = LogisticRegression(solver='liblinear')
```

# Cross validation 

```python
# cross validation
model_kf = KFold(10)
```

```python
result_validation = cross_val_score(logreg, X_train, y_train, cv= model_kf)
```

```python
result_validation
```

```python
print(np.mean(result_validation))
```

# Fit the model to the train data

```python
# fit model on dataset train
logreg.fit(X_train, y_train.ravel())
```

# prediction

```python
# prediction
y_pred = logreg.predict(X_test)
```

```python
y_pred
```

# Comparing the predicted value with the actual value

```python
# compare peredict and actual
compare =pd.DataFrame({'actual': y_test.flatten(),
          'predict' : y_pred.flatten()})
```

```python
compare
```

# Accuracy

```python
print("Accuracy : ",metrics.accuracy_score(y_test, y_pred))
```

```python
# Drawing plot
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred)
plt.plot(fpr, tpr, label='data1')
plt.legend(loc=4)
# plt.legend(loc='best')
# plt.legend(loc=6)
plt.show()
```

```python
# Drawing plot
y_pred_proba = logreg.predict_proba(X_test)[ : : ,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label='data1')
plt.legend(loc='best')
plt.show()
```
